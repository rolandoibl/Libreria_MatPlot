El primer paso a hacer es importar nuestras librerías de numpy y matplotlib a nuestro proyecto para poder trabajar libremente con una amplia gama de funciones. 

```python
"""El comando "as" nos permite importar nuestra librería con
otro nombre más corto para hacer más rápida y sencilla nuestra programación
"""
import matplotlib.pyplot as plt    #librería de matplot
import numpy as np                 #librería de numpy
```

### Función plot()

La función plot() nos va imprimir los datos que le pasemos en dos listas, estos datos nos imprimiran una línea en la que el primer argumento será los valores en x y el segundo será los valores en y. El método show() nos mostrará la gráfica que queremos graficar.

```python
x = [2,4,6,8]
y = [2,4,6,9]
plt.plot(x,y)
plt.show()
```

## Configuración de la gráfica 

Los rasgos muy importantes en una gráfica son el título de la gráfica y los ejes coordenados X y Y, por lo que veremos a continuación como modificar dichos parámetros.

```python
plt.plot(x,y)
plt.title("Temperatura")  #La función title() nos cambia el título de nuestra gráfica
plt.xlabel("Tiempo [s]")  #La función xlabel() nos modifica la etiqueta en el eje de las X
plt.ylabel("T [°C]")      #La función ylabel() nos modifica la etiqueta en el eje de las Y
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(x,y,"ro")            #Se cambia a círculos con nuestro 3er párametro el cuál la "r" es red 
plt.show()                    
```

Si cambiamos el valor de "ro" cambiar la forma de nuestra línea, por lo cual nos podemos guiar por la siguiente tabla para saber la salida que deseamos.

Color: red = r ;      blue = b ;       green = g ;      yellow = y ;

Forma: dash = -- ;    square = s ;     triangle = ^ ;   circle = o ;

Si no cambiamos ningún parametro obtendremos una línea.

### Gráfico de dispersión

```python
x = np.linspace(0,20,100)         #Función de numpy,linspace(inicio, final, intervalo), crea un arreglo de valores de un rango dado
y = np.random.randn(100)          #Función de numpy, random.randn(cantidad), crea un arreglo de la cantidad pedida con número aleatorios
plt.scatter(x,y)                  #Función para hacer gráficos de disper
plt.show()
```

### Múltiples gráficos

#### Funciones en un mismo gráfico

```python
x = np.linspace(0,20,100)
y = np.sin(x)       #Función de numpy para obtener la función seno
z = np.cos(x)       #Función de numpy para obtener la función coseno

plt.plot(x,y,label='Sin x')      #El tercer parámetro en plot() es la etiqueta de nuestra gráfica la cual nos sirve
plt.plot(x,z,label='Cos x')      #para poder diferenciar una función de otra
plt.legend(loc='upper right')
plt.show()
```

#### Funciones en dos diferentes gráficos, misma figura

```python
x = np.linspace(0,20,100)
y = np.sin(x)
z = np.cos(x)

plt.figure(figsize=(15,5))

#En cada función tendremos que crear un subplot con diferentes parámetros que nos indican la posición en la figura
plt.subplot(131) #Figura con una fila y tres columnas, activo primer subgráfico
plt.plot(x,y,'g-')
plt.title('Funcion 1')
plt.ylabel('Amplitud')

plt.subplot(132) #Figura con una fila y tres columnas, activo segundo subgráfico
plt.plot(x,z,'b.')
plt.title('Funcion 2')
plt.ylabel('Amplitud')

plt.subplot(133) # Figura con una fila y tres columnas, activo tercer subgráfico
plt.plot(y,x,'r')
plt.title('Funcion 3')
plt.ylabel('Amplitud')

plt.show()
```

#### Gráficos en diferentes ventanas

```python
x = np.linspace(0,20,100)
y = np.sin(x)
z = np.cos(x)

#Al agregar un número a la función figure() nos indica que es otra ventana con diferente gráfico
plt.figure(1)  #Ventana 1
plt.plot(x,y)
plt.title('Sin x')
plt.show()

plt.figure(2)  #Ventana 2
plt.plot(x,z)
plt.title('Cos x')
plt.show()
```

### Histograma 

```python
dist = np.random.randn(200) 
plt.hist(dist) #La funcion hist() nos grafica el histograma de un arreglo de datos
plt.show()
```

### Gráfico semilogarítmico

```python
xpb = [1000,2000,3000,4000,5000,6000]
angpb = [-23.49,-50.70,-77.06,-96.19,-111.6,-123.55]
ganpb = [1,0.851,0.7307,0.5925,0.4615,0.3846]

plt.figure(1)
plt.semilogx(xpb,angpb,'r-')   #Funcion para escala logaritmica en x
plt.show()

plt.figure(2)
plt.semilogy(ganpb,xpb,'g.')   #Funcion para escala logaritmica en y
plt.show()

plt.figure(3)
plt.loglog(xpb,xpb,'b^')       #Funcion para escala semilogaritmica en ambos ejes
plt.show()
```

## Guardar figuras en imágenes

```python
x = np.linspace(0,20,100)
y = np.sin(x)

#Creamos nuestro gráfico de forma usual
plt.plot(x,y)
plt.show()

#El comando savefig() es el que nos guarda la figura y tiene como parámetro el nombre con el que deseamos guardar y su extension
plt.savefig("sin.eps")                  #Guardar en formato de vector .eps
plt.savefig("seno.png",dpi=300)         #Guardar en formato imagen .png, con una resolucion de 300DPI
```
